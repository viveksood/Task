```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv('Expander_data.csv')
raw_data
```

```python
raw_data.describe(include='all')
```

```python
raw_data.columns
```

```python
data=raw_data.drop(['Timestamp','By-pass Valve Position (%)','Run Status','Suction Temperature (F)','Discharge Temperature (F)'],axis=1)
data.head()
```

```python
sns.distplot(data['Discharge Pressure (psig)'])
```

```python
plt.scatter(data['Suction Pressure (psig)'], data['Discharge Pressure (psig)'], s=10)
```

# Regression

```python
targets=data['Discharge Pressure (psig)']
inputs=data.iloc[:,0:3]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(inputs)
```

```python
scaled_inputs=scaler.transform(inputs)
scaled_inputs
```

# train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,test_size=0.2,random_state=365)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat=reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('targets')
plt.ylabel('predictions')
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title("Residual plot",size=18)
```

# TEST THE MODEL

# import your test set here

```python

y_hat_test=reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=.2)
plt.xlabel('targets')
plt.ylabel('predictions')
plt.show()
```

```python
pred_df=pd.DataFrame(y_hat_test,columns=['Predictions'])
pred_df.head()
```
